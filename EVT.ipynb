# Extreme Value Theory (EVT)

## EVT Approach to Estimate Worst-Case BTC Drawdown

### Step 1: Define Drawdown

The drawdown at time $t$ is defined as: $$D_t = \frac{P_t - \max(P_0,\ P_1,\ \ldots,\ P_t)}{\max(P_0,\ P_1,\ \ldots,\ P_t)}$$
This is the percentage decline from the historical peak. We're interested in the maximum drawdowns over time.

### Step 2: Prepare the Drawdown Time Series

```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
# Load BTC data
btc = yf.download("BTC-USD", start="2013-01-01", end="2025-01-01")['Close']

# Compute drawdowns
running_max = btc.cummax()
drawdown = (btc - running_max) / running_max
drawdown.name = 'Drawdown'
```

```python
drawdown.plot()
```
