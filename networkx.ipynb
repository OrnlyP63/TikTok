```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Load Zachary's Karate Club graph
G = nx.karate_club_graph()

# Compute positions using the Fruchterman-Reingold force-directed algorithm
pos = nx.spring_layout(G, seed=42)

# Color nodes by their club affiliation
club_colors = {'Mr. Hi': 'skyblue', 'Officer': 'orange'}
node_colors = [club_colors[G.nodes[i]['club']] for i in G.nodes]

# Draw the graph
plt.figure(figsize=(10, 10))
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=800, edgecolors='black', linewidths=1.5)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

plt.title("Zachary's Karate Club Network", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```
