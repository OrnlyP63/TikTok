# Correlation: ตรวจหารูปแบบเชิงเส้นตรง
1. Standard Deviation: วัดการกระจายของข้อมูลจากค่าเฉลี่ย

พิจารณา ตัวแปรสุ่ม $X$ $$\sigma_X = \sqrt{Var(X)} = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \bar{x})^2}$$

2. Covariance (Cov): วัดความสัมพันธ์ของสองตัวแปร

พิจารณาตัวแปรสุ่ม $X$ และ $Y$ $$Cov(X, Y) = \frac{1}{n}\sum_{i=1}^n(x_i-\bar{x})(y_i-\bar{y})$$
## Pearson Correlation: 

$$\rho_{X, Y} = \frac{Cov(X, Y)}{\sigma_X\sigma_Y}$$

```python
import numpy as np
import matplotlib.pyplot as plt

#Set Seaborn style
import seaborn as sns
sns.set(style="whitegrid", palette="deep", context="notebook")
```

```python
n = 1000
x = np.random.uniform(-1, 1, n) # input /  features
y = x + 0.1 * np.random.randn(n) # output / labels
```

```python
x_bar = np.mean(x) # mean x
y_bar = np.mean(y) # mean y

sigma_x = np.sqrt(np.sum((x - x_bar) ** 2) / n)
sigma_y = np.sqrt(np.sum((y - y_bar) ** 2) / n)

sigma_x, sigma_y
```

```python
cov_xy = np.sum((x - x_bar) * (y - y_bar)) / n

cov_xy
```

```python
corr = cov_xy / (sigma_x * sigma_y)
corr
```

```python
# Plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, color='blue', alpha=0.6)  # removed fontsize
plt.title(f'Correlation: {corr:.3f}', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.grid(True)
plt.show()
```
